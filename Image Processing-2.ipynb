```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Question 1
# To read image from disk, we use
# cv2.imread function, in below method,
img = cv2.imread("SanDiego.jpg")
A = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.title("original color image")
plt.imshow(A)
plt.show()
print(A.shape)
```

```python
# # Question 2
G = cv2.cvtColor(A, cv2.COLOR_BGR2GRAY)
print(G.shape)
# cv2.imshow('grayscale image', G)
# cv2.waitKey(0)
plt.axis("off")
plt.title("grayscale image")
plt.imshow(G,cmap='gray')
plt.show()
```

```python
# Question 3
print("Size of A", A.shape)
print("Size of G", G.shape)

# Size of A (1712, 2288, 3)
# Size of G (1712, 2288)
# In RGB images there are three color channels and has three dimensions,therefore we have (1712 * 2288 * 3)=
# 11751168pixels.
# while grayscale images are single-dimensional,therefore we have (1712 * 2288) =3917056 pixels.
```

```python
# # Question 4
print(type(G))
print(f'Before changing {G.dtype}')
# changing the data type of numpy array using astype() method
B = G.astype(np.float64)
# type of array after changing
print(f'\nAfter changing {B.dtype}')
C=B/255.0
print(G)
print(C)
plt.axis("off")
plt.title("Normalized grayscale")
plt.imshow(B,cmap='gray')
plt.show()

```

```python
# # Question 5

# Question a
D1=pow(C,0.25)
print(D1)

# Question b
D2=pow(C,0.5)
print(D2)
D3=pow(C,1.0)
print(D3)
D4=pow(C,1.5)
print(D4)

# Question c
Titles =["0.25", "0.5", "1.0", "1.5"]
images =[D1, D2, D3, D4]
count = 4
for i in range(count):
    plt.axis("off")
    plt.subplot(2, 2, i + 1)
    plt.title("contrast variation "+Titles[i])
    plt.imshow(images[i],cmap='gray')
plt.axis("off")
plt.show()

# Question d
# By adjusting the factor you can adjust the contrast of the image.
#While a factor of 1 gives normalised image back.
#While a factor of 0 makes all the pixel values 1, which is a white image.
#While the pixel values are in the range[0,1], as the power factor increases , the pixel value gets smaller and the
# images tend to get towards white.

```

```python
# Question e
#These are unrecognizable.

D5=pow(C,4.0)
D6=pow(C,5.0)
D7=pow(C,6.0)
D8=pow(C,8.0)


Titles2 =["4.0", "5.0", "6.0", "8.0"]
images2 =[D5, D6, D7, D8]
count2 = 4
for i in range(count2):
    plt.axis("off")
    plt.subplot(2, 2, i + 1)
    plt.title("contrast variation "+Titles2[i])
    plt.imshow(images2[i],cmap='gray')
plt.axis("off")
plt.show()
```

```python
# Question f
# When you save the image you need to consider the data type. To save the normalized image as jpg, 
# you need to scale the normalized values to integer range 
# (such as [0, 255]) or use image format that supports floating point format.
print(D3)
img10=D3*255
print(img10)
print("Size of the saved image ",img10.shape)
cv2.imwrite('Normalized 1.0.jpg', img10)

# reading the saved image back
#If you save 2D ndarray to a file and read it again with cv2.imread(), 
#it will be read as 3D ndarray in which each color is the same value.
#It does not automatically read in as a two-dimensional array.

img1 = cv2.imread("Normalized 1.0.jpg")
print("Size of loaded image ",img1.shape)
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
print("Size of grayscale image " ,img2.shape)
img3 = img2.astype(np.float64)
img4=img3/255.0
print(img4)


Titles1 =["Saved Image", "Loaded Image"]
images1 =[D3,img4]
count1 = 2
for i in range(count1):
    plt.axis("off")
    plt.subplot(1, 2, i + 1)
    plt.title(Titles1[i])
    plt.imshow(images1[i],cmap='gray')
plt.axis("off")
plt.show()

```
